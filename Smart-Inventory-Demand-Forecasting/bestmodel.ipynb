```python
from pathlib import Path
import json
import subprocess
import sys

project_root = Path.cwd()
subprocess.run([sys.executable, "run_training.py"], check=True)

metrics_path = project_root / "models" / "model_comparison_metrics.json"
with open(metrics_path, "r", encoding="utf-8") as f:
    metrics = json.load(f)

print("Deployed model:", metrics.get("deployed_model"))
print("Validation split:", metrics.get("validation_split", "n/a"))
for name, vals in metrics.get("models", {}).items():
    print(f"{name}: R2={vals['r2']:.4f}, RMSE={vals['rmse']:.4f}, MAE={vals['mae']:.4f}")
```
